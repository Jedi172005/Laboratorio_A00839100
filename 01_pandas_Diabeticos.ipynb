```python
## Se puede apreciar en el boxplot de edad que la mayoria de los pacientes tienen unas edades de entre 24 años y 41 años. A pesar de esto el rango normal va desde 
##alrededor de los 21 años hasta alrededor de los 65 años. Se pueden apreciar unos cuantos outliers, con personas superando la edad normal del rango con edades arriba 
## de 65 años. 
```

```python
sns.boxplot(diabetes, x="Outcome", y="Glucose")
plt.title("Glucosa de acuerdo a pacientes con diabetes")
```

```python
## Podemos ver en este boxplot y la comparacion que, para las personas que no tienen diabetes, existen de un rango de alredeor de 55 de glucosa hasta 175 de glucosa
## Para las personas que si tienen diabetes, el rango va desde alrededor de 80 de glucosa hasta 200 de glucosa. La media de las personas que no tienen diabetes
## esta entre los 90 y 125 de glucosa, mientras que el de las personas con diabetes este entre los 120 de glucosa hasta los 170 de glucosa. Solo existen outliers
## en el caso de las personas que no tienen diabetes, superando los niveles normales del rango de glucosa.
```

```python
sns.boxplot(diabetes, x="Outcome", y="BMI")
plt.title("BMI de acuerdo a pacientes con diabetes")
```

```python
## Podemos ver que el BMI de las personas sin diabetes va desde los 19 hasta alrededor de 47, mientras que las personas con diabetes tienen un rango de BMI desde
## los 23 hasta alrededor de los 51. La media de las personas sin diabetes es de un BMI entre 25 y 35, mientras que el de las personas con diabetes esta entre 31 y 37.
## Finalmente, podemos ver que existen outliers en ambos casos, igual inferiores y superiores. 
```

```python
variables_numericas = diabetes.select_dtypes(include='number')
matriz_correlation = variables_numericas.corr().round(2)
matriz_correlation
```

```python
## Se puede notar ciertas relaciones fuertes, como puede ser Age con el numero de pregnancies, al igual que el glucosa y BMI con el outcome. Otras que resaltan
## son las de insulina con la glucosa, con un valor de 0.33.
```

```python
sns.heatmap(matriz_correlation, annot=True)
plt.title('Correlacion entre caracteristicas (Features)')
```
